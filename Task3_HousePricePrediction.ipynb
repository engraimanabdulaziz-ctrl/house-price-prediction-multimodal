```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import torch
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image

# =========================
# 1. Load CSV
# =========================
csv_path = "/content/housing_dataset/socal2.csv"
img_dir = "/content/housing_dataset/socal2/socal_pics"

df = pd.read_csv(csv_path)
print("Data shape:", df.shape)
print("Columns:", df.columns.tolist())

# Target variable
y = df["price"].values

# Drop unused cols
X = df.drop(columns=["price", "image_id", "street"])

# Separate categorical & numeric
cat_cols = ["citi"]   # categorical
num_cols = ["n_citi", "bed", "bath", "sqft"]

# Preprocess tabular
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols)
    ]
)

X_tab = preprocessor.fit_transform(X)
print("Tabular features shape:", X_tab.shape)

# =========================
# 2. Prepare images (random pairing)
# =========================
all_images = sorted(os.listdir(img_dir))  # get all filenames
n = min(len(df), len(all_images))         # match only min length
df = df.iloc[:n]                          # trim df if needed
y = y[:n]
X_tab = X_tab[:n]

print("Rows:", len(df), "Images:", len(all_images))

transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

class HouseImageDataset(Dataset):
    def __init__(self, img_files, img_dir, transform=None):
        self.img_files = img_files
        self.img_dir = img_dir
        self.transform = transform
    def __len__(self):
        return len(self.img_files)
    def __getitem__(self, idx):
        fname = self.img_files[idx]
        path = os.path.join(self.img_dir, fname)
        img = Image.open(path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img

dataset = HouseImageDataset(all_images[:n], img_dir, transform)
loader = DataLoader(dataset, batch_size=16, shuffle=False)

# =========================
# 3. Extract Image Features (ResNet18)
# =========================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
resnet = models.resnet18(pretrained=True)
resnet.fc = nn.Identity()
resnet = resnet.to(device)
resnet.eval()

image_features = []
with torch.no_grad():
    for batch in loader:
        batch = batch.to(device)
        feats = resnet(batch)
        image_features.append(feats.cpu().numpy())

image_features = np.vstack(image_features)
print("Image features shape:", image_features.shape)

# =========================
# 4. Combine + Train model
# =========================
X_combined = np.hstack([X_tab, image_features])
print("Combined shape:", X_combined.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.2, random_state=42
)

model = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f" MAE: {mae:.2f}")
print(f" RMSE: {rmse:.2f}")
```
